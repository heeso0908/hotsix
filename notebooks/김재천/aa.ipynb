```python
df = pd.read_csv("../../data/df_type1_파생변수추가.csv")
```

```python
import pandas as pd

# 1. 파일 불러오기


# --- 방법 1: value_counts(normalize=True) 사용 (가장 권장됨) ---
# normalize=True를 사용하면 각 값의 비율(클래스 비율)을 바로 구해줍니다.
defect_rates = df['Defect_Status'].value_counts(normalize=True)
print("Product Type 1 DATA")
print(f"전체 불량 or 양품 비율:\n{defect_rates}\n")

defect_rate = defect_rates[1] # 값이 0(불량)인 비율
print(f"불량률 (Defect_Status == 1): {defect_rate:.2%}")

# --- 방법 2: 전체 행의 수와 불량 행의 수를 직접 카운트해서 나누기 ---
total_count = len(df)
defect_count = len(df[df['Defect_Status'] == 1])
normal_count = len(df[df['Defect_Status'] == 0])

defect_rate_manual = defect_count / total_count
print(f"\n총 데이터 개수: {total_count}개")
print(f"불량 데이터 개수: {defect_count}개")
print(f"양품 데이터 개수: {normal_count}개")
print(f"불량률: {defect_rate_manual:.2%}")
```
