```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 

df = pd.read_csv(
    'DieCasting_Quality_Raw_Data.csv',
    header=[0, 1]
)
display(df.head(10))
```

```python
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
df.columns = df.columns.get_level_values(1).str.strip()
```

```python
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
```

```python
print(f"데이터 로드 결과: {df.shape[0]}행 × {df.shape[1]}열")
```

```python
df.value_counts()
```

```python
missing_counts = df.isnull().sum()
missing_cols = missing_counts[missing_counts > 0]

missing_cols.sort_values(ascending=False)
```

```python
missing_df = pd.DataFrame({
    'missing_count': missing_cols,
    'missing_ratio': df[missing_cols.index].isnull().mean() * 100
})

missing_df.sort_values('missing_count', ascending=False)
```

```python
print(df.columns.tolist())
```

```python
factory_cols = missing_cols.index

df[factory_cols] = (
    df.groupby('Product_Type')[factory_cols]
      .transform(lambda x: x.fillna(x.median()))
)
```

```python
print(f" 잔여 결측값: {df.isnull().sum().sum()}건")
```
